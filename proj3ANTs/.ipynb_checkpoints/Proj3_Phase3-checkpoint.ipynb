## Coding Solution
Use ```super().parentmethod()``` to inherit parent class methods

````python
class Water(Place):
    """Water is a place that can only hold waterproof insects."""

    def add_insect(self, insect):
        """Add an Insect to this place. If the insect is not waterproof, reduce
        its health to 0."""
        # BEGIN Problem 10
        "*** YOUR CODE HERE ***"
        super().add_insect(insect)
        if not insect.is_waterproof:
            insect.reduce_health(insect.health)
        # END Problem 10
````

```python
    def action(self, gamestate):
        """A Bee's action stings the Ant that blocks its exit if it is blocked,
        or moves to the exit of its current place otherwise.

        gamestate -- The GameState, used to access game state information.
        """
        # Extra credit: Special handling for bee direction
        if self.scared: # When scared, change direction, bee moves to place.entrance (backward)
            destination = self.place.entrance
            self.scared_turns -= 1
        else:
            destination = self.place.exit # By default, bee moves to place.exit (forward)

        if self.slowed:
            self.slow_turns -= 1 
            if self.slow_turns == 0: # Slow when turns >= 1
                self.slowed = False # When turns counter goes 0, no longer being slowed
            if gamestate.time % 2 == 0 and self.health > 0 and destination is not None:
                self.move_to(destination) # Move only when time is even
            elif self.scared:
                self.scared_turns += 1 # If scared and bee can't move, we need to fix scared turns by adding one
            else:
                pass
        else:
            if self.blocked():
                self.sting(self.place.ant)
            elif self.health > 0 and destination is not None:
                self.move_to(destination)

        if self.scared_turns == 0:
            self.scared = False # After confirming no longer being scared, set scared to false
```

````python
class Bee(Insect):
    """A Bee moves from place to place, following exits and stinging ants."""

    name = 'Bee'
    damage = 1
    is_waterproof = True
    # OVERRIDE CLASS ATTRIBUTES HERE

    slowed = False
    scared = False

    slow_turns = 0
    scared_turns = 0
    was_scared = False

    def sting(self, ant):
        """Attack an ANT, reducing its health by 1."""
        ant.reduce_health(self.damage)

    def move_to(self, place):
        """Move from the Bee's current Place to a new PLACE."""
        self.place.remove_insect(self)
        place.add_insect(self)

    def blocked(self):
        """Return True if this Bee cannot advance to the next Place."""
        # Special handling for NinjaAnt
        # BEGIN Problem Optional 1
        return self.place.ant is not None
        # END Problem Optional 1

    def action(self, gamestate):
        """A Bee's action stings the Ant that blocks its exit if it is blocked,
        or moves to the exit of its current place otherwise.

        gamestate -- The GameState, used to access game state information.
        """
        if self.scared:
            destination = self.place.entrance
            self.scared_turns -= 1
        else:
            destination = self.place.exit

        # Extra credit: Special handling for bee direction
        if self.slowed:
            self.slow_turns -= 1
            if self.slow_turns == 0:
                self.slowed = False
            if gamestate.time % 2 == 0 and self.health > 0 and destination is not None:
                self.move_to(destination)
            elif self.scared:
                self.scared_turns += 1
            else:
                pass
        else:
            if self.blocked():
                self.sting(self.place.ant)
            elif self.health > 0 and destination is not None:
                self.move_to(destination)

        if self.scared_turns == 0:
            self.scared = False

    def add_to(self, place):
        place.bees.append(self)
        Insect.add_to(self, place)

    def remove_from(self, place):
        place.bees.remove(self)
        Insect.remove_from(self, place)

    def slow(self, length):
        """Slow the bee for a further LENGTH turns."""
        # BEGIN Problem EC
        "*** YOUR CODE HERE ***"
        self.slowed = True
        self.slow_turns += length
        # END Problem EC

    def scare(self, length):
        """
        If this Bee has not been scared before, cause it to attempt to
        go backwards LENGTH times.
        """
        # BEGIN Problem EC
        "*** YOUR CODE HERE ***"
        if self.was_scared:
            return
        else:
            self.was_scared = True
            self.scared_turns = length
            self.scared = True
        # END Problem EC
````
